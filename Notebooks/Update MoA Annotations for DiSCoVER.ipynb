```python
import pandas as pd
from io import StringIO
import nbtools
import numpy as np

##### Before DiSCoVER
def split_discover_dataframe(df, min_score=0):
    # This is super inefficient for larger DataFrames, but let's worry about efficiency later

    out = pd.DataFrame(columns=['moa','GDSC','CTRP','CCLE','drug'])

    for index, row in df.iterrows():
        dataset = row['drug'].split('_')[0].lower()
        drug = row['drug'].split('_')[1].lower()

        if dataset == 'gdsc':
            out.loc[drug,'GDSC'] = row['score']
        elif dataset == 'ctrp':
            out.loc[drug,'CTRP'] = row['score']
        elif dataset == 'ccle':
            out.loc[drug,'CCLE'] = row['score']
        out.loc[drug,'drug'] = row['drug']
        out.loc[drug,'moa'] = row['moa']

    #     df['database'] = [drug.split('_')[0].upper() for drug in df['drug']]
    #     df['drug'] = [drug.split('_')[1].lower() for drug in df['drug']]
    #     df = df[df['score']>0]  # It feels inneficient to do this last, but Pandas complaints if I do it earlier.
    #     out = df[~df['drug'].duplicated(keep='first')].drop('score', axis=1,inplace=False)
    #     out['Drug'] = np.unique([drug.split('_')[1].lower() for drug in df['drug']])
    return out

sign_to_letter = {
    1:"+",
    -1:"-",
    '1.0':"+",
    '-1.0':"-",
    'nan':'.',
    '0.0':'.',
    '0':'.',
    0:'.',
}

annotation = {
    "motesanib":"VEGFR inhibitor, receptor tyrosine kinase inhibitor, PDGFR inhibitor, Kit inhibitor, Ret receptors inhibitor, Cell proliferation inhibitor",
    'cetuximab':"EGFR inhibitor",
    'gw-2580':"Tyrosine Kinase inhibitor",
    'dabrafenib':'BRAF inhibitor',
    'cyclopamine':'glioma-associated oncogene inhibitor',
    'rucaparib':'PARP inhibitor',
    'plx-4720':'BRAF inhibitor',
}

def supporting_evidence(row):
    # assuming only three columns
    return sign_to_letter[str(np.sign(row[0]))]+sign_to_letter[str(np.sign(row[1]))]+sign_to_letter[str(np.sign(row[2]))]


def rank_drugs_discover(df):
    df['score'] = df.drop(['moa','drug'],axis=1,inplace=False).mean(axis=1,skipna=True).round(3)
    df['evidence'] = df.drop(['moa','score'],axis=1).apply(supporting_evidence, axis=1)
    return df.sort_values(by=['score'],ascending=False,axis=0)

def make_clickable(url):
    to_display = url.split('=')[1]
    return '<a href="{}" target="_blank">{}</a>'.format(url, to_display)



@nbtools.build_ui(
        name="Make drug table",
        description='Create an easy to read table based on the output from DiSCoVER.',
        parameters={
                    "GP_csv": {  "name": "DiSCoVER output CSV file",
                                 "type": "file",
                                 "kinds": ["csv","CSV"],
                                 "description":"The output from DiSCoVER",
                            },
                   "number_of_drugs": {"default":10},
                   "output_var": {"default":"drug_df"},
                })
def make_table(GP_csv:'URL of the CSV output from DiSCoVER',
               number_of_drugs:'How many drugs to report'=10,
              ):
    uio = nbtools.UIOutput(name='Example Output',status='Running!')
    
    uio.status = 'Fetching csv file...'
    # Get the job number and name of the file
#     temp = GP_csv.split('/')
    # programatically access that job to open the file
#     GP_csv = eval('job'+temp[5]+'.get_file("'+temp[6]+'")')
#     drug_df = pd.read_csv(StringIO(GP_csv.read()),index_col=0)
#     drug_df['drug'] = drug_df.index
    df = pd.read_csv(GP_csv)

    df = split_discover_dataframe(df=drug_df)
    df = rank_drugs_discover(df)
    df['Pubchem link']=pd.Series(["https://pubchem.ncbi.nlm.nih.gov/#query={}".format(drug) for drug in df.index.values], index=df.index)
    df['Manual annotation']=pd.Series([annotation[drug] if drug in annotation.keys() else 'N/A' for drug in df.index.values], index=df.index)
    
    annotation
    styled_df = df.head(number_of_drugs).style.format({'Pubchem link': make_clickable})
    
    uio.status = 'Formatting csv file...'
    display(styled_df)
    return df
```

```python
def find_synonyms_in_row(of, row, force_lowercase=True):
    to_return = False
    for sym in row['synonyms']:
        if force_lowercase:
            to_return = to_return or (sym.lower()==of)
        else:
            to_return = to_return or (sym==of)
    return to_return

def find_synonyms_in_df(of, df):
    return df.apply(lambda x: find_synonyms_in_row(of,x), axis=1)    

def count_synonyms_in_df(of, df):
    return sum(df.apply(lambda x: find_synonyms_in_row(of,x), axis=1))

def check_for_synonyms(df):
    all_good = True
    list_of_repeated_drugs = []
    for drug in df.index:
        if count_synonyms_in_df(of=drug, df=df)>1:
            all_good = False
            list_of_repeated_drugs.append(drug)
    if all_good:
        return 0
    else:
        return list_of_repeated_drugs
```

```python
def merge_non_overlapping(df,row_to_keep, row_to_drop):
    merged_synonyms = list(set(df.loc[row_to_keep,'synonyms'] + df.loc[row_to_drop,'synonyms']))
    df.loc[row_to_keep] = df.loc[row_to_keep].combine_first(df.loc[row_to_drop])
    df.loc[row_to_keep,'synonyms'] = merged_synonyms
    df = df.drop(row_to_drop)
    return df
```

```python
def merge_repeated(df,drug_to_keep):
    merged_synonyms = df.loc[drug_to_keep,'synonyms']
    temp_drug = df.loc[[drug_to_keep]].drop('synonyms',axis=1).drop_duplicates()
    temp_drug['synonyms'] = [merged_synonyms.iloc[0]]
    temp_drug.name = drug_to_keep
    df = df.drop(drug_to_keep)
    df = df.append(temp_drug)
    return df.sort_index()
```

```python
def merge_superset_repeated(df,drug_to_keep,which_one):
    repeated_rows = df.loc[drug_to_keep]
    df = df.drop(drug_to_keep)
    df = df.append(repeated_rows.iloc[which_one])
    return df.sort_index()
```

```python
def duplicated(drugs,drug_to_keep):
    drugs = merge_repeated(df=drugs,drug_to_keep=drug_to_keep)
    if len(drugs.loc[[drug_to_keep]]) == 1:
        display(drugs.loc[[drug_to_keep]])
        print(f'Repeated rows for {drug_to_keep} were sucessfully removed!')
    else:
        display(drugs.loc[[drug_to_keep]])
        print(f'Repeated rows for {drug_to_keep} are still present :/')
    return drugs
```

```python
def superset(drugs,drug_to_keep,index_to_keep):
    drugs = merge_superset_repeated(drugs,drug_to_keep,index_to_keep)
    if len(drugs.loc[[drug_to_keep]]) == 1:
        display(drugs.loc[[drug_to_keep]])
        print(f'Repeated rows for {drug_to_keep} were sucessfully removed!')
    else:
        display(drugs.loc[[drug_to_keep]])
        print(f'Repeated rows for {drug_to_keep} are still present :/')
    return drugs
```

```python
def merge(drugs,drug_to_keep,drug_to_drop):
    drugs = merge_non_overlapping(drugs,drug_to_keep,drug_to_drop)
    display(drugs.loc[[drug_to_keep]])
    if drug_to_drop in drugs.index:
        print(f"Drug {drug_to_drop} is somehow still in index.")
        display(drugs.loc[[drug_to_drop]])
    else:
        print(f"Drug {drug_to_drop} was sucessfully removed fron the index!")
    return drugs
```

# Combine Targets

```python
drugs = pd.read_csv('moa_annotations/CCLE_CTRP_GDSC_drugs_merged.csv',index_col=0)
drugs
```

```python
def combine_targets(row):
    combined_targets = []
    if row['ccle_target']==row['ccle_target']:
        combined_targets.append(row['ccle_target'])
    if row['gdsc_target']==row['gdsc_target']:
        for targ in row['gdsc_target'].split('; '):
            combined_targets.append(targ)
    if row['ctrp_target']==row['ctrp_target']:
        for targ in row['ctrp_target'].split('; '):
            combined_targets.append(targ)
    if len(combined_targets)==0:
        combined_targets.append('NaN')
    row['targets'] = list2str(list(set(combined_targets)))
    return row
```

```python
drugs = drugs.apply(lambda x: combine_targets(x), axis=1)
drugs
```
