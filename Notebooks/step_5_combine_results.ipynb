```python
all_drugs = screen_results[['drug_name','sbi_id','pubchem_cid']]
all_drugs
```

```python
def find_sbi_id(text,all_drugs=all_drugs):
    text = str(text)
    text_found = all_drugs[all_drugs['drug_name'].str.contains(text, regex=False,na=False)]
    # text_found['pubchem_cid'] ## should return pubchem_cid
    return list(text_found['sbi_id'])
```

```python
find_sbi_id('linsitinib')
```

```python

combined_sheet['SBI'] = combined_sheet['Name'].apply(find_sbi_id)
combined_sheet
```

```python
all_drugs_from_exp = pd.DataFrame()
temp = combined_sheet
temp['sbi_id'] = temp['SBI']
col_order = temp.columns
print(col_order)
for index,row in combined_sheet.iterrows():
    row = row[col_order]
    sbi_list = row['SBI']
    if len(sbi_list) == 0:
        row['sbi_id'] = 'NaN'
    elif len(sbi_list) == 1:
        row['sbi_id'] = sbi_list[0]
    else:
        if COMPARE:
            # there are multiple SBIs, so row will be a dataframe, not a series
            old_row = row
            row = pd.DataFrame()
            for curr_sbi in sbi_list:
                old_row['sbi_id'] = curr_sbi
                row = row.append(old_row)
        else:
            row['sbi_id'] = sbi_list[0] 
    all_drugs_from_exp = all_drugs_from_exp.append(row)
all_drugs_from_exp.reset_index(drop=True,inplace=True)
all_drugs_from_exp = all_drugs_from_exp[col_order]
all_drugs_from_exp
```

```python
effective_exp = all_drugs_from_exp[all_drugs_from_exp.apply(lambda x: any([x['G3'],x['G4'],x['SHH']]), axis=1)]
effective_exp = effective_exp.reset_index(drop=True)
effective_exp
```
